Unlike the a priori method, the ECLAT method is not based on the calculation of confidence and lift, therefore the ECLAT method is based on the calculation of the support conjunctions of the variables

https://pypi.org/project/pyECLAT/

```python
import pandas as pd
import ast
```

```python
dataset = pd.read_csv("Retail_Transactions_Dataset.csv")

dataset.head()
```

```python
product_list = dataset["Product"].apply(ast.literal_eval).tolist()

products_dataframe = pd.DataFrame(product_list)

products_dataframe
```

```python
from pyECLAT import ECLAT

eclat_instance = ECLAT(data=products_dataframe,
                       verbose=True)  #verbose=True to see the loading bar
```

```python
eclat_instance.df_bin  #generate a binary dataframe, that can be used for other analyzes.
eclat_instance.uniq_  #a list with all the names of the different items
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit(min_support=0.001,
                                                           min_combination=2,
                                                           max_combination=2,
                                                           separator=' & ',
                                                           verbose=True)
```
