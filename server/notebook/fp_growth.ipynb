```python
import pandas as pd
import ast
```

```python
dataset = pd.read_csv("Retail_Transactions_Dataset.csv")

dataset.head()
```

```python
product_list = dataset["Product"].apply(ast.literal_eval).tolist()

product_list
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(product_list).transform(product_list)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import fpgrowth

support_Two_items = fpgrowth(df, min_support=0.001, use_colnames=True)

support_two_items = support_Two_items[support_Two_items['itemsets'].apply(
    lambda x: len(x) == 2)]
display(support_two_items)
```
